# Espectogramas

El espectrograma es una herramienta básica de representación que se utiliza para el análisis de las señales eléctricas, de comunicaciones, y cualquier señal audiovisual en su contenido frecuencial.

Es una representación en tres dimensiones, temporal, frecuencial y amplitud de la distribución de energía de una señal.



```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
import librosa.display
import os
import pandas as pd  
```

```python
def create_spectrogram_test(filename, name):
    plt.interactive(False)
    clip, sample_rate = librosa.load(filename, sr=None)
    clip = librosa.resample(clip, sample_rate, 8000)
    clip = librosa.resample(clip, sample_rate, 8000)
    sample_rate = 8000
    print(len(clip))
    plt.figure(figsize=(35,20))
    plt.subplot(3, 1, 1)
    plt.title('Monophonic ' + name)
    librosa.display.waveplot(clip, sr=sample_rate)
    plt.show() 
```

```python
create_spectrogram_test('chords/Amaj/100.wav', 'Amaj')
```

```python
create_spectrogram_test('chords/Fmaj/1.wav', 'Fmaj')
```

```python
create_spectrogram_test('chords/G#dim/19940.wav', 'G#dim')
```

```python
create_spectrogram_test('chords/G#dim/3651.wav', 'G#dim')
```
